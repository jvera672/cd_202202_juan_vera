```python
import math

def distancia(x, y):
  return math.sqrt(x**2 + y**2)
```

```python
cont = 0
puntosLado = 10000
paso = 1 / (puntosLado - 1)

i = 0
while i <= 1:
  j = 0
  while j <= 1:
    if distancia(i,j) <= 1:
      cont += 1
    j += paso
  i += paso
print("Puntos:", cont)
print("Estimaci贸n de pi:", ((cont/puntosLado**2)*4))
```

```python
import random as rnd
import plotly.express as px

cont = 0
xValues = []
yValues = []
numberOfPoints = 10000
for i in range(numberOfPoints):
  x = rnd.random()
  y = rnd.random()
  if distancia(x,y) <= 1:
    xValues.append(x)
    yValues.append(y)
    cont += 1
print("Puntos:", cont)
print("Estimaci贸n de pi:", ((cont/numberOfPoints)*4))
fig = px.scatter(x=xValues, y=yValues, width=400, height=400)
fig.show()
```
