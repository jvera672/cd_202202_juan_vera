```python
import pandas as pd
```

```python
filePath = 'https://foscalinternacional.com/datasets/eventos_dengue_2020.csv'
df = pd.read_csv(filePath)
df
```

```python
df.dtypes
```

```python
dfSexArray = df.sexo_
numSexo = []
for value in dfSexArray:
  if(value == 'F'):
    numSexo.append(1)
  else:
    numSexo.append(0)
df['newSexo'] = numSexo
print(numSexo)
```

```python
dfAgeValue = df.edad_
dfAgeType = df.uni_med_
numEdad = []

for i in range(len(dfAgeValue)):
  if(dfAgeType[i] == 1):
    numEdad.append(float(dfAgeValue[i]))
  elif(dfAgeType[i] == 2):
    numEdad.append(dfAgeValue[i]/12)
  else:
    numEdad.append(dfAgeValue[i]/365)
df['newEdad'] = numEdad
```

```python
dfSalud = df.tip_ss_
numSalud = []
for value in dfSalud:
  if(value == 'Contributivo'):
    numSalud.append(0)
  elif(value == 'Especial'):
    numSalud.append(1)
  elif(value == 'Subsidiado'):
    numSalud.append(2)
  elif(value == 'Particular'):
    numSalud.append(3)
  else:
    numSalud.append(4)
df['newSalud'] = numSalud
```

```python
dfConducta = df.conducta
numConducta = []
for value in dfConducta:
  if(value == '1. Ambulatorio'):
    numConducta.append(0)
  elif(value == '2. Hospitalizacion piso'):
    numConducta.append(1)
  elif(value == '3. UCI'):
    numConducta.append(2)
  else:
    numConducta.append(3)
df['newConducta'] = numConducta
```

```python
df.drop(columns=['ID', 'fec_not','año', 'sexo_', 'uni_med_', 'cod_pais_o', 'cod_dpto_o', 'cod_mun_o', 'tip_ss_', 'cod_ase_', 'fecha_contagio', 'inicio_sintomas', 'fec_hos_', 'condicion_final', 'fecha_defuncion', 'fecha_nacimiento', 'certif_defuncion', 'conducta'])
```
