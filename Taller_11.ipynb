```python
import statsmodels.api as sm
results_summary = result.summary()

results_as_html = results_summary.tables[1].as_html()
table = pd.read_html(results_as_html, header=0, index_col=0)[0]
table
```

```python
#table = result.summary2().tables[1]
ageCoef = table['coef'].values[0]
sibSpCoef = table['coef'].values[1]
parchCoef = table['coef'].values[2]
```

#Con sexo

```python
x = train.iloc[:,[5,6,7,11]]
y = train.Survived
logit_model = sm.Logit(y,x)
result=logit_model.fit()
print(result.summary())
```

```python
results_summary = result.summary()

results_as_html = results_summary.tables[1].as_html()
table = pd.read_html(results_as_html, header=0, index_col=0)[0]
table
```

```python
ageCoefSex = table['coef'].values[0]
sibSpCoefSex = table['coef'].values[1]
parchCoefSex = table['coef'].values[2]
sex2CoefSex = table['coef'].values[3]
```

#Cálculos sin variable sexo

```python
fn = 0
tp = 0
fp = 0
tn = 0
for index, row in test.iterrows():
    prob = 1 / (1 + math.e ** (-1* (ageCoef*row['Age']+sibSpCoef*row['SibSp']+parchCoef*row['Parch'])))
    if(prob >= 0 and prob < 0.5):
      trunc = 0
    elif(prob > 0.5 and prob <= 1.0):
      trunc = 1
    trueVal = row['Survived']

    if(trueVal == 1 and trunc == 1):
      tp += 1
    elif(trueVal == 0 and trunc == 1):
      fp += 1
    elif(trueVal == 1 and trunc == 0):
      fn += 1
    elif(trueVal == 0 and trunc == 0):
      tn += 1
print("Falsos Negativos: " + str(fn))
print("Falsos Positivos: " + str(fp))
print("Verdaderos Negativos: " + str(tn))
print("Verdaderos Positivos: " + str(tp))
```

```python
p = tp/(tp+fp)
r = tp/(fn+tp)
print("Precisión: " + str(p*100))
print("Exhaustividad: " + str(r*100))
```

#Cálculo con variable sexo

```python
fn = 0
tp = 0
fp = 0
tn = 0
for index, row in test.iterrows():
    prob = 1 / (1 + math.e ** (-1* (ageCoefSex*row['Age']+sibSpCoefSex*row['SibSp']+parchCoefSex*row['Parch']+sex2CoefSex*row['Sex2'])))
    if(prob >= 0 and prob < 0.5):
      trunc = 0
    elif(prob > 0.5 and prob <= 1.0):
      trunc = 1
    trueVal = row['Survived']

    if(trueVal == 1 and trunc == 1):
      tp += 1
    elif(trueVal == 0 and trunc == 1):
      fp += 1
    elif(trueVal == 1 and trunc == 0):
      fn += 1
    elif(trueVal == 0 and trunc == 0):
      tn += 1
print("Falsos Negativos: " + str(fn))
print("Falsos Positivos: " + str(fp))
print("Verdaderos Negativos: " + str(tn))
print("Verdaderos Positivos: " + str(tp))
```

```python
p = tp/(tp+fp)
r = tp/(fn+tp)
print("Precisión: " + str(p*100))
print("Exhaustividad: " + str(r*100))
```
