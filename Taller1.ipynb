```python
import pandas as pd
```

```python
filePath = 'https://foscalinternacional.com/datasets/eventos_dengue_2020.csv'
df = pd.read_csv(filePath)
df
```

```python
df.dtypes
```

```python
from datetime import datetime
df['fec_not2'] = pd.to_datetime(df['fec_not'], format='%d/%m/%Y')
df['fecha_contagio2'] = pd.to_datetime(df['fecha_contagio'], format='%d/%m/%Y')
df['inicio_sintomas2'] = pd.to_datetime(df['inicio_sintomas'], format='%d/%m/%Y')
df['fec_hos_2'] = pd.to_datetime(df['fec_hos_'], format='%d/%m/%Y')
df['fecha_nacimiento2'] = pd.to_datetime(df['fecha_nacimiento'], format='%d/%m/%Y')
```

```python
import pyproj

P = pyproj.Proj(proj='utm', zone=18, ellps='WGS84', preserve_units=True)
```

```python
df['x'], df['y'] = P(df['longitud'], df['latitud'])
df
```

```python
#7.062918322155907, -73.08607298204963 alcaldía
```

```python
import math
xPoint, yPoint = P(-73.08607298204963, 7.062918322155907)
xValues = df['x']
yValues = df['y']
distanceValues = []
for i in range(len(xValues)):
  distanceValues.append(math.sqrt((abs(xPoint - xValues[i]))**2 + (abs(yPoint - yValues[i]))**2))
print(xPoint)
print(distanceValues)
```

```python
df['dist'] = distanceValues
```
